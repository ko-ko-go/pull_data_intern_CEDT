```python
# ==========================================
# 1. SETUP & IMPORTS
# ==========================================
import pandas as pd
import re
import ast
import numpy as np  # เพิ่ม numpy เพื่อจัดการค่า infinity
from bs4 import BeautifulSoup
import gspread
from oauth2client.service_account import ServiceAccountCredentials

# ==========================================
# 2. FUNCTIONS
# ==========================================
def clean_html(text):
    if pd.isna(text) or text == "" or str(text).lower() == 'none':
        return ""
    soup = BeautifulSoup(str(text), "html.parser")
    clean_text = soup.get_text(separator=' ')
    clean_text = re.sub(r'\s+', ' ', clean_text).strip()
    return clean_text

def clean_tags(tag_data):
    if isinstance(tag_data, list):
        if len(tag_data) == 0:
            return ""
        tag_names = [str(t.get('tagName', '')) for t in tag_data if isinstance(t, dict)]
        return ", ".join(filter(None, tag_names))
    
    if pd.isna(tag_data) or tag_data == "" or tag_data == "[]":
        return ""
    
    if isinstance(tag_data, str):
        try:
            tags_list = ast.literal_eval(tag_data)
            if isinstance(tags_list, list):
                tag_names = [str(t.get('tagName', '')) for t in tags_list if isinstance(t, dict)]
                return ", ".join(filter(None, tag_names))
        except:
            return str(tag_data)
    return str(tag_data)

# ==========================================
# 3. FETCH DATA
# ==========================================
print("กำลังดึงข้อมูลจาก API...")


data = response.json()
df = pd.json_normalize(data['items'])
print(f"ดึงข้อมูลสำเร็จ: {len(df)} แถว")

# ==========================================
# 4. CLEANING & CALCULATION (เพิ่มส่วนคำนวณที่นี่)
# ==========================================

# 4.1 Clean HTML
cols_to_clean = ['description', 'requirements']
for col in cols_to_clean:
    if col in df.columns:
        df[col] = df[col].apply(clean_html)

# 4.2 Clean Tags
if 'tags' in df.columns:
    df['tags'] = df['tags'].apply(clean_tags)

# 4.3 Rename Columns
new_names = {
    'company.companyNameTh': 'company',
    'compensationType.compensationType': 'compensation_Type'
}
df = df.rename(columns=new_names)

# 4.4 Convert Types (แปลงเป็นตัวเลขเพื่อคำนวณ)
df['compensationAmount'] = pd.to_numeric(df['compensationAmount'], errors='coerce').fillna(0)
df['quota'] = pd.to_numeric(df['quota'], errors='coerce').fillna(1) # กันหารด้วย 0
df['inStudentDraftCount'] = pd.to_numeric(df['inStudentDraftCount'], errors='coerce').fillna(0)

# -------------------------------------------------------
# [ใหม่] สร้างคอลัมน์วิเคราะห์การแข่งขัน
# สูตร: จำนวนคนสนใจ / จำนวนรับ
# -------------------------------------------------------
df['competition_rate'] = df['inStudentDraftCount'] / df['quota']

# ปัดทศนิยม 2 ตำแหน่ง และจัดการค่า Infinity (กรณี Quota เป็น 0)
df['competition_rate'] = df['competition_rate'].round(2)
df['competition_rate'] = df['competition_rate'].replace([np.inf, -np.inf], 0)

# 4.5 Sort Values
df = df.sort_values(by='compensationAmount', ascending=False)

# ==========================================
# 5. SELECT COLUMNS (จัดเรียงคอลัมน์ใหม่)
# ==========================================
keep_cols = [
    'company', 
    'title', 
    'quota', 
    'inStudentDraftCount',
    'competition_rate',   # <--- แทรกตรงนี้ตามที่ขอ
    'compensationAmount', 
    'compensation_Type', 
    'tags',
    'workingCondition', 
    'description', 
    'requirements',
    'officeName', 
    'officeAddressLine1', 
    'officeAddressLine2'
]

# กรองคอลัมน์
existing_cols = [c for c in keep_cols if c in df.columns]
df = df[existing_cols]

# ดูตัวอย่างผลลัพธ์
print(df[['title', 'quota', 'inStudentDraftCount', 'competition_rate']].head())

# ==========================================
# 6. EXPORT & UPLOAD
# ==========================================
df.to_csv('cleaned_data-v2.csv', index=False, encoding='utf-8-sig')

try:
    print("กำลังอัปโหลดขึ้น Google Sheets...")
    
    scope = ["https://spreadsheets.google.com/feeds", "https://www.googleapis.com/auth/drive"]
    creds = ServiceAccountCredentials.from_json_keyfile_name('key.json', scope) 
    client = gspread.authorize(creds)
    
    sheet_name = "intern_cedt" 
    sheet = client.open(sheet_name).sheet1

    # Safety: Convert all to string
    df_export = df.fillna('').astype(str)

    data_to_upload = [df_export.columns.values.tolist()] + df_export.values.tolist()

    sheet.clear()
    sheet.update(values=data_to_upload)
    
    print(f"✅ สำเร็จ! อัปเดตข้อมูลขึ้น Google Sheet '{sheet_name}' เรียบร้อยแล้ว")

     
    # sheet_name = "intern-cho" 
    # sheet = client.open(sheet_name).sheet1

    # # Safety: Convert all to string
    # df_export = df.fillna('').astype(str)

    # data_to_upload = [df_export.columns.values.tolist()] + df_export.values.tolist()

    # sheet.clear()
    # sheet.update(values=data_to_upload)
    
    # print(f"✅ สำเร็จ! อัปเดตข้อมูลขึ้น Google Sheet '{sheet_name}' เรียบร้อยแล้ว")

except Exception as e:
    print(f"❌ Error: {e}")
```
